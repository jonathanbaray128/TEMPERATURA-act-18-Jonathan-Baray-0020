#### Fórmulas de Conversión
Para movernos entre estas dos escalas, utilizamos las siguientes relaciones matemáticas:De Celsius a Fahrenheit: Multiplicamos los grados Celsius por $1.8$ (o $9/5$) y sumamos $32$.$$°F = (°C \times 1.8) + 32$$De Fahrenheit a Celsius: Restamos $32$ a los grados Fahrenheit y luego dividimos el resultado entre $1.8$.$$°C = \frac{°F - 32}{1.8}$$
#### Ejemplos Resueltos Ejemplo A:
Ejemplos ResueltosEjemplo A: Convertir 25°C (Temperatura ambiente) a °FSustitución: Tomamos la fórmula $°F = (25 \times 1.8) + 32$.Multiplicación: Primero resolvemos el paréntesis: $25 \times 1.8 = 45$.Suma: Sumamos los 32 constantes: $45 + 32 = 77$.Resultado: 77°F.

```python
# Funcion para convertir de celsius a fahrenheit
def convertirCF():
    celsius = float(input("\nIngrese los grados Celsius (°C): "))
    fahrenheit = (celsius * 1.8) + 32
    print(f"---> Resultado: {celsius}°C equivalen a {fahrenheit:.2f}°F")

# Funcion para convertir de fahrenheit a celsius
def convertirFC():
    fahrenheit = float(input("\nIngrese los grados Fahrenheit: "))
    celsius = (fahrenheit - 32) * 5/9
    print(f"---{fahrenheit}°F es igual a {celsius:.2f}°C ---")

# Funcion principal del menu
def menu():

    while True:
        print("\n--- MENÚ DE CONVERSIONES ---")
        print("1. Convertir grados Centígrados a Fahrenheit")
        print("2. Convertir grados Fahrenheit a Centígrados")
        print("3. Salir")

        opcion = input("Seleccione una opción (1-3): ")

        if opcion == "1":
            convertirCF()
        elif opcion == "2":
            convertirFC()
        elif opcion == "3":
            print("Saliendo del programa... ¡Hasta luego!")
            print("Jonathan Baray NL = 06")
            break
        else:
            print("Opción no válida. Intente de nuevo.")

# Punto de entrada del programa
if __name__ == "__main__":
    menu()
```
